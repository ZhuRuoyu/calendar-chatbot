```python
#!pip install sentence_transformers

from sentence_transformers import SentenceTransformer
```

```python
model_pretrained = 'bert-base-nli-mean-tokens'
```

```python
sentences_prebuild = [
    "Creating an event", #0
    "retrieving event of today", #1
    "retrieving event of tomorrow", #2
    "retrieving event of next month", #3
    "add one new event",  #4
    "check event attendees", #5
    "delete a event",#6
    "retrieving event of next week"
    ]
```

```python
sentences_testing = [

"What is today's highlighted event?",
"Can you tell me the main happening of the day?",
"I'd like to know the event that's taking place today, please.",
"What is the significant occurrence for today?",
"Could you inform me about the event scheduled for today?",
"I'm curious about the day's noteworthy event. What is it?",
"Can you provide information on the event happening today?",
"I'm interested in knowing the special event of the day.",
"Could you share details about today's prominent event?",
"Who is in the meeting?",
"Am I free for next week?",
"what am I going to do next week?",
]

```

```python
model = SentenceTransformer(model_pretrained)
```

```python
vecs_prebuild = model.encode(sentences_prebuild)
vecs_testing = model.encode(sentences_testing)
```

```python
vecs_prebuild.shape #(8, 768)
len(vecs_testing) #12
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
for i in range(len(vecs_testing)):
  print('TEST SENTENCE: ',sentences_testing[i])
  vector = vecs_testing[i]
  cosine_score = cosine_similarity([vector], vecs_prebuild)
  #print(cosine_score)
  index = np.argmax(cosine_score, axis = 1)[0]
  print(index)
  print('PRE BUILD SENTENCE:',sentences_prebuild[index])
  print('-------- next line----------')
```
